# Appendix - Quartile Groups

## Background	
During the project, each feature that was categorical with a large (>5) number of categories was split into four categories using the following method:
1. For each category, create a list of all of the prices of cards that fall into that category.
2. Find the median for each list of card prices.
3. Make a list of all of the median card prices.
4. Split the median card prices into 4 quartile groups.  The lowest 25% of median card prices were labeled as the Q1 group.  The categories between 25% and 50% were labeled the Q2 group.  Between 50% and 75% was the Q3 group and the top 25% became the Q4 group.
